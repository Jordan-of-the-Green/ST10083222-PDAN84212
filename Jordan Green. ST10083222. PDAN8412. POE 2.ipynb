### **2. Explanation of Why the Data Set is Appropriate for Logistic Regression**

#### **2.1. What is Logistic Regression**

***Based on a given data set of independent variables, logistic regression calculates the likelihood that an event will occur, such as voting or not(IBM, 2024).***

***Predictive analytics and classification frequently use this kind of statistical model, also called a logit model. Because the result is a probability, the dependent variable has a range of 0 to 1. A logit transformation is performed to the odds in logistic regression, which are the probability of success divided by the probability of failure. This is sometimes referred to as the natural logarithm of odds or the log odds.(IBM, 2024)***
***

#### **2.2. Appropriateness: Logistic Regression**

***Binary Classification Problem: The primary target variable, pha, indicates whether an object is a potentially hazardous asteroid (PHA). The values of this variable are 0 (not a PHA) and 1 (is a PHA). Since logistic regression is optimized for binary outcomes, it is a perfect fit for this classification problem.***  
***

***Feature Relationship: Logistic regression assumes a linear relationship between the log-odds of the outcome and the predictor variables(IBM, 2024). Logistic regression can yield significant insights if the relationship between the predictors and the target variable can be accurately represented by a linear model, or if the relationship can be translated into a linear form. The dataset is suitable for logistic regression analysis since it contains a number of numerical and categorical variables (such as diameter, rotation period, and orbital characteristics) that may affect whether an object is classed as a PHA.***
***

***Robustness to Overfitting: Logistic regression is less prone to overfitting than more complicated models, particularly when lower sample sizes are involved. Logistic regression offers a reliable solution that strikes a compromise between model complexity and generalization, notwithstanding the possibility of a small number of observations in the dataset.(IBM, 2024)***
***

***Ease of Implementation: As a realistic option for preliminary analysis, logistic regression is simple to apply and computationally efficient(IBM, 2024). For comparison with more intricate algorithms in later studies, it provides a solid foundational paradigm. ## **2) Logistic Regression Model:**  

### **Preparing Data for Model Training:**  

```python
# ***************************************************************************************
#    Title: How to implement Logistic Regression from scratch with Python
#    Author: AssemblyAI
#    Date Code/Created: 09 October 2024
#    Date Retrieved: 14 September 2022
#    Code version: 1
#    Availability: https://youtu.be/YYEJ_GUguHw?si=skj3DkhFByAf8Ifj
#
# ***************************************************************************************

# Imports for standard linear algebra
import numpy as np  

# Sigmoid function introduces non-linearity by mapping values to the range (0, 1)
def sigmoid(x: np.ndarray) -> np.ndarray:
    return 1 / (1 + np.exp(-x))  # Formula: 1 / (1 + e^-x)

# Create a LogisticRegressionModel class for binary classification
class LogisticRegressionModel:
    
    # Learning rate (lr) controls the step size in parameter updates.
    # A smaller lr ensures smaller updates and slower convergence, 
    # while n_iters controls the number of iterations over the training data.
    def __init__(self, lr: float = 0.001, n_iters: int = 1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None  # Weights for each feature
        self.bias = None  # Bias term

    # Fit method to train the model with gradient descent
    def fit(self, X: np.ndarray, y: np.ndarray):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)  # Initialize weights
        self.bias = 0  # Initialize bias

        # Gradient descent optimization
        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            predictions = sigmoid(linear_pred)

            # Gradients for weights and bias
            dw = (1 / n_samples) * np.dot(X.T, (predictions - y))
            db = (1 / n_samples) * np.sum(predictions - y)

            # Update weights and bias
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    # Method to make predictions
    def predict(self, X: np.ndarray) -> np.ndarray:
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        # Classify as 0 or 1 based on probability threshold 0.5
        class_pred = np.where(y_pred > 0.5, 1, 0)
        return class_pred
```

### **Generating predictions on test data:**  

```python
# ***************************************************************************************
#    Title: Machine Learning - Train/Test
#    Author: W3Schools
#    Date Code/Created: 11 October 2024
#    Date Retrieved: 03 September 2024
#    Code version: 1
#    Availability: https://www.w3schools.com/python/python_ml_train_test.asp
#
# ***************************************************************************************

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Setting up x and y from your DataFrame
x = df_clean_pandas.drop(['pha'], axis=1)  # x is the data without 'pha'
y = df_clean_pandas['pha']  # y is the 'pha' column

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.3,
    shuffle=True,
    stratify=y,
    random_state=42
)

# Check for non-numeric columns in x_train
print("Data types in x_train:")
print(x_train.dtypes)

# Convert categorical columns to numeric using LabelEncoder
label_encoder = LabelEncoder()

for col in x_train.columns:
    if x_train[col].dtype == 'object':  # If the column contains categorical data
        x_train[col] = label_encoder.fit_transform(x_train[col])
        # Apply the same transformation to x_test
        if col in x_test.columns:
            x_test[col] = label_encoder.transform(x_test[col])

# Align x_test with x_train after encoding
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)

# Convert to NumPy arrays
x_train = np.array(x_train)
x_test = np.array(x_test)

# Ensure y_train is a NumPy array
y_train = np.array(y_train)

# Instantiate LogisticRegressionModel
classifier = LogisticRegressionModel(lr=0.001, n_iters=1000)

# Fit the model using the training data
classifier.fit(x_train, y_train)

# Generate predictions on the test data (x_test)
predictions = classifier.predict(x_test)
```

## **4) Evaluation of the Model:**  

### **Method to calculate the accuracy of the model:**  

```python
# Method to calculate and return the accuracy of the model
def accuracy(predictions: np.ndarray, y_test: np.ndarray) -> float:
    """Calculate accuracy as the proportion of correct predictions."""
    correct_predictions = np.sum(predictions == y_test)
    total_predictions = len(y_test)
    return correct_predictions / total_predictions

# Method to evaluate the model based on accuracy
def evaluate_accuracy(acc: float) -> str:
    """Provide feedback on the model's accuracy."""
    if acc >= 0.90:
        return "Good accuracy! The model performs well."
    elif acc >= 0.75:
        return "Moderate accuracy. The model is acceptable but could be improved."
    else:
        return "Low accuracy! The model needs improvement."

# Calculate the accuracy of predictions
acc = accuracy(predictions, y_test)

# Display the accuracy
print(f"Accuracy: {acc:.2%}")  # Format accuracy as a percentage

# Display feedback on the accuracy
feedback = evaluate_accuracy(acc)
print(feedback)

# Note: Although the accuracy score returned is high (e.g., 99%),
# an imbalanced dataset can skew results, making this score potentially misleading.
```

```python
# Imports
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)

# Calculate the F1 score for the predictions
# The F1 score is the harmonic mean between precision and recall
f1 = f1_score(predictions, y_test, average="weighted")

# Display the F1 score
print("F1 Score:", f1)
```

### **Overall performance assessment:**  

```python
# ***************************************************************************************
#    Title: How to create a seaborn correlation heatmap in Python?
#    Author: GeeksforGeeks
#    Date Code/Created: 12 September 2024
#    Date Retrieved: 06 October 2024
#    Code version: 1
#    Availability: https://www.geeksforgeeks.org/how-to-create-a-seaborn-correlation-heatmap-in-python/
#
# ***************************************************************************************

# Import necessary libraries
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the confusion matrix
cm = confusion_matrix(y_test, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)

# Create a heatmap using the correct variable 'cm'
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Not NEO', 'NEO'], 
            yticklabels=['Not NEO', 'NEO'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Extract True Negatives, False Positives, False Negatives, and True Positives
tn, fp, fn, tp = cm.ravel()

# Calculate Precision, Recall, and F1 Score manually
precision = tp / (tp + fp)  # Precision: TP / (TP + FP)
recall = tp / (tp + fn)     # Recall (Sensitivity): TP / (TP + FN)
f1 = 2 * (precision * recall) / (precision + recall)  # F1 Score: Harmonic mean of Precision and Recall

# Display the results
print(f"True Negatives (TN): {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")
print(f"True Positives (TP): {tp}")
print()
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print()

# Interpretation of the confusion matrix values
if fn > 0:
    print(f"The model has made {fn} False Negatives, indicating Type II errors (missed positive cases).")
else:
    print("There are no False Negatives, indicating that all actual positive cases were correctly identified!")

if fp > 0:
    print(f"The model has made {fp} False Positives, indicating Type I errors (incorrect positive predictions).")
else:
    print("There are no False Positives, indicating that all negative cases were correctly identified!")

# Overall performance assessment
if tp > 0 and tn > 0:
    print("Overall, the model appears to be performing well.")
else:
    print("Overall, the model may need improvement, as there are issues with correctly identifying positive or negative cases.")

print()    
```
